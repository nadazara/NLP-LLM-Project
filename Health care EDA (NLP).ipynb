```python
train_data  = pd.read_csv('/content/doctor_reviews_200.csv')
train_data.head(10)
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Split dataset to train
x = train_data['Doctor_Review']
y = train_data['Class']
```

```python

X_train , X_test , Y_train ,Y_test = train_test_split(x, y , random_state = 42 , test_size =.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizor = TfidfVectorizer()
X_train_vec = vectorizor.fit_transform(X_train)
X_test_vec = vectorizor.transform(X_test)

model = RandomForestClassifier()
model.fit(X_train_vec , Y_train)
```

```python
prediction = model.predict(X_test_vec)
print(prediction)
```

```python
data['Classfication'] = model.predict(vectorizor.transform(data['Correct_Feedback']))
```

```python
data['Classfication'].value_counts()
```

```python
structured_data = pd.DataFrame({
  'Patient Name':data['Patient Name'].values,
  'Age': data['Clean Age'].values,
  'Gender':data['Correct_Gender'].values,
  'Diagnosis':data['Correct_Diagnosis'].values,
  'Hospital Name':data['Hospital Name'].values,
  'Admission Date':data['Admission Date'].values,
  'Discharge Date': data['Discharge Date'].values,
  'Medication':data['Correct_Medication'].values,
  'Sport':data['Correct_Sport'].values,
  'Feedback': data['Correct_Feedback'].values,
  'Classfication' : data['Classfication'].values
})
```
